# Plot position data

```python
import pickle
import datetime
import matplotlib.pyplot as plt

with open('checkpoints-lap-2.pickle', 'rb') as f:
  data = pickle.load(f)

x_val = [x[0] for x in data[1]]
y_val = [-x[1] for x in data[1]]

fig, ax = plt.subplots()
ax.plot(x_val, y_val)
plt.title("Goodwood Circuit")
fig.savefig("checkpoints-lap-2.png")
plt.show()
```

```python
min(x_val), max(x_val)
```

```python
dist = len(x_val) / (1/60) * 1e-5
print(f"distance in km: {dist}")
```

```python
laptime = str(datetime.timedelta(seconds=len(x_val) * 1/60))
print(f"race lap time: {laptime}")
```
